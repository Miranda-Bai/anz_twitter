```python
import pandas as pd
```

```python
df1=pd.read_excel("data/twscrape1_original.xlsx")
df2=pd.read_excel("data/twscrape2_original.xlsx")
```

```python
df1.describe()
```

```python
df2.describe()
```

```python
# Combine DataFrames vertically using concat
combined_df = pd.concat([df1, df2], ignore_index=True)

# Drop duplicates based on the "id_str" and "date" columns, keeping the first occurrence
result_df = combined_df.drop_duplicates(subset=["id_str","date"], keep="first")

result_df.shape
```

```python
result_df.head()
```

```python
result_df[result_df['hashtages'] == '[]']
```

```python
drop_ad = result_df[result_df['hashtages'] != '[]']
# Reset the index 
drop_ad.reset_index(drop=True, inplace=True)
drop_ad.shape
```

```python
drop_ad.head()
```

```python
drop_ad[drop_ad['hashtages'] == '[]']
```

```python
drop_ad[drop_ad['lang'] != 'en']
```

```python
# only keep lang='en'
en_df = drop_ad[drop_ad['lang'] == 'en']
en_df.shape
```

```python
en_df.to_excel("./data/preprocessed.xlsx")
```
